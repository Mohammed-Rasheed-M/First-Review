PAIR WISE ALIGNMENT 

The match parameters are:

CODE  DESCRIPTION
x     No parameters. Identical characters have score of 1, otherwise 0.
m     A match score is the score of identical chars, otherwise mismatch
      score.
d     A dictionary returns the score of any pair of characters.
c     A callback function returns scores.

The gap penalty parameters are:

CODE  DESCRIPTION
x     No gap penalties.
s     Same open and extend gap penalties for both sequences.
d     The sequences have different open and extend gap penalties.
c     A callback function returns the gap penalties.

```python
X = df[['INDEX','LENGTH']].copy()
```

```python
from sklearn.mixture import GaussianMixture
n_clusters = 20
gmm_model = GaussianMixture(n_components=n_clusters)
gmm_model.fit(X)
```

```python
cluster_labels = gmm_model.predict(X)
X = pd.DataFrame(X)
X['cluster'] = cluster_labels
```

```python
for k in range(0,n_clusters):
    data = X[X["cluster"]==k]
    plt.scatter(data["INDEX"],data["LENGTH"])
    plt.title("Clusters Identified by Guassian Mixture Model")    
    plt.ylabel("LENGTH")
    plt.xlabel("INDEX")
    plt.show()
```

```python
from Bio import SeqIO

# Replace 'filename.fasta' with the actual filename of the FASTA file you downloaded
sequences = list(SeqIO.parse('sequence.fasta', 'fasta'))
```

```python
from Bio import pairwise2

# Replace sequences[0] and sequences[1] with the actual SeqRecord objects you want to align
alignment = pairwise2.align.globalxx(sequences[0].seq, sequences[1].seq)

# Print the alignment
print(pairwise2.format_alignment(*alignment[0]))
```

```python
from Bio.Align import MultipleSeqAlignment

# Perform pairwise sequence alignment between all pairs of sequences
pairwise_alignments = []
for i in range(len(sequences)):
    for j in range(i+1, len(sequences)):
        alignment = pairwise2.align.globalxx(sequences[i].seq, sequences[j].seq)
        pairwise_alignments.append(alignment[0])

# Sort the pairwise alignments by score
pairwise_alignments.sort(key=lambda x: x.score, reverse=True)

# Create a consensus sequence by aligning the sequences progressively
consensus = sequences[0].seq
for alignment in pairwise_alignments:
    aligned_seq = alignment.seqB if alignment.seqA == consensus else alignment.seqA
    consensus = MultipleSeqAlignment([consensus, aligned_seq]).consensus(threshold=0.6, ambiguous='N')

# Print the final alignment
alignment = MultipleSeqAlignment(sequences)
print(alignment)
```
